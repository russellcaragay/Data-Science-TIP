<h3>Checking Null Entries in Training Features</h3>

```python
#Making the training features' column to list
col_trainf = training_features.columns.tolist()

#Creating a list of total null per column
col_trainfnullsums = (training_features.isnull().sum()).to_list()

#Creating a list of average null per column
col_trainfnullpercent = (training_features.isnull().mean().round(5)*100).to_list()

#Creating a dataframe of the total sum and percent of null each column
null_trainf_Df = pd.DataFrame({'Column': col_trainf,'Null Entries': col_trainfnullsums,'Null Percent': col_trainfnullpercent})
```

```python
null_trainf_Df
```

<h3>Checking Null Entries in Test Features</h3>

```python
#Making the test features' column to list
col_testf = test_features.columns.tolist()

#Creating a list of total null per column
col_testfnullsums = (test_features.isnull().sum()).to_list()

#Creating a list of average null per column
col_testfnullpercent = (test_features.isnull().mean().round(5)*100).to_list()

#Creating a list of unique features per column
col_testfunique = test_features.nunique().to_list()

#Creating a dataframe of the total sum and percent of null each column
null_testf_Df = pd.DataFrame({'Column': col_testf,'Null Entries': col_testfnullsums,'Null Percentage': col_testfnullpercent,'Unique Entries': col_testfunique})
```

```python
null_testf_Df
```

<h3>Filling Null Entries in Training Features and Test Features</h3>

```python
#Checking the columns that has Null values from the dataframe created earlier
trainf_null_columns = null_trainf_Df[null_trainf_Df['Null Entries'] != 0].Column.to_list()

#Filling null entries in the training features
for col in trainf_null_columns:
        training_features[col]=training_features[col].fillna(training_features[col].mode()[0])
```

```python
#training_features where null entries are removed
training_features
```

```python
#Checking the columns that has Null values from the dataframe created earlier
testf_null_columns = null_testf_Df[null_testf_Df['Null Entries'] != 0].Column.to_list()

#Filling null entries in the training features
for col in testf_null_columns:
        test_features[col]=test_features[col].fillna(test_features[col].mode()[0])
```

```python
#training_features where null entries are removed
test_features
```

```python
#Creating Category Columns
cat_columns = training_features.select_dtypes(include=['object']).columns.to_list()
for col in cat_columns:
    training_features[col] = training_features[col].astype('category')
    training_features[col] = training_features[col].cat.codes
    test_features[col] = test_features[col].astype('category')
    test_features[col] = test_features[col].cat.codes
```

```python
cat_columns
```

<h3>Creating the Model</h3>

```python
#Importing Necessary Libraries for modelling
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import GradientBoostingClassifier,VotingClassifier
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,AdaBoostRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import svm
from xgboost import XGBClassifier  
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#Using the GradientBoostingClassifier and Checking the Accuracy
X_train, X_eval, y_train, y_eval = train_test_split(
    training_features,
    training_labels['h1n1_vaccine'],
    test_size=0.33,
    shuffle=True,
    stratify=training_labels['h1n1_vaccine'],
    random_state=27
)

clf2 = GradientBoostingClassifier(learning_rate=.5,random_state=27)
clf2.fit(X_train, y_train)  
y_pred2 = clf2.predict(X_train)
print("Accuracy:",metrics.accuracy_score(y_train, y_pred2))
```

```python
#Using the GradientBoostingClassifier and Checking the Accuracy
clf3 = LogisticRegression(penalty="l2",solver="newton-cg", C=10)
clf3.fit(X_train, y_train)  
y_pred3 = clf3.predict(X_train)
print("Accuracy:",metrics.accuracy_score(y_train, y_pred3))
```

```python
#Initializing the training features and labels
X= training_features
y= training_labels

X_train, X_eval, y_train, y_eval = train_test_split(
    X,
    y,
    test_size=0.33,
    shuffle=True,
    stratify=y,
    random_state=27
)
```

```python
#Using the XGBOOST classifier and Checking the Accuracy
from sklearn import svm
from xgboost import XGBClassifier  
clf1 = XGBClassifier()  
clf1.fit(X_train, y_train)  
y_pred = clf1.predict(X_train)
print("Accuracy:", metrics.accuracy_score(y_train, y_pred))
```

```python
#Creating all the classifiers used earlier
clf1 = XGBClassifier() 
clf2 = GradientBoostingClassifier(learning_rate=.5,random_state=27)
clf3 = LogisticRegression(penalty="l2",solver="newton-cg", C=10)

#Since the clasifiers tested earlier do not support multiple output, we will use the MultiOutputClassifier of sklearn
mult_classifier = MultiOutputClassifier(estimator=VotingClassifier(estimators=[('ab', clf1), ('gb', clf2), ('lr', clf3)], voting='soft'))
```

```python
#fitting the multiclassifier created in the previous cell
mult_classifier.fit(X_train,y_train)
```

<h3>Testing</h3>

```python
test = mult_classifier.predict_proba(test_features)
test
```
